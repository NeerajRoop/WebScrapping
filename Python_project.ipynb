```python
# source -https://www.w3resource.com/python-exercises/web-scraping/web-scraping-exercise-23.php
#https://bit.ly/2NyxdAG
from bs4 import BeautifulSoup
import requests
import re


# Download IMDB's Top 50 data
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

movies = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]

imdb = []
urls=[]
# Store each item into dictionary (data), then put those into a list (imdb)
for index in range(0,50):
    # Seperate movie into: 'place', 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    f=str(index+1)
    url1 = "https://www.imdb.com"+links[index]
    response = requests.get(url1)
    soup = BeautifulSoup(response.text, 'lxml')
    
    genre = str(soup.select('div.see-more.inline.canwrap')[1].text).replace('Genres:','').replace('|','')
    genre=genre.strip().replace(' ','').replace('\n','').split('\xa0')
    desc= soup.select('div.summary_text')[0].contents
    desc=str(str(desc[0]).strip()).lower()
    place = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "place": place,
            "rating": ratings[index],
            "genre" : genre,
            "description" : desc
           }
    imdb.append(data)
    urls.append(url1)

i=0
main=[]
for item in imdb:
    main.append(["imdb",item['place'], item['movie_title'], str(round(float(item['rating']),1)),item['description'],item['genre']])
print("Done")
```

```python
# source -https://www.w3resource.com/python-exercises/web-scraping/web-scraping-exercise-23.php
#https://bit.ly/2NyxdAG
from bs4 import BeautifulSoup
import requests
import re

# Download IMDB's Top 50 data
url = 'https://www.rottentomatoes.com/top/bestofrt/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

movies = soup.select('td')
ratings = [b.attrs.get('data-value') for b in soup.select('td span[class=tMeterScore]')]
links = [a.attrs.get('href') for a in soup.select('td a')]

rt = []
url=[]
# Store each item into dictionary (data), then put those into a list (imdb)
for index in range(48,248):
    # Seperate movie into: 'place', 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie
    rt.append(movie_title)
    
for i in range(48,98):
    url.append(links[i])

i=0
while(i<len(rt)):
    b=rt[i+2].split()
    b.pop()
    c=''
    for t in b:
        c=c+" "+t
    rt[i+2]=c
    a=list(rt[i+1])
    a.pop()
    c=''
    for t in a:
        c=c+t
    c=float(int(c)/10)
    rt[i+1]=c
    i+=4
genres=[]
desc=[]
for u in url:
    urls = 'https://www.rottentomatoes.com'+u
    response = requests.get(urls)
    soup = BeautifulSoup(response.text, 'lxml')
    gres = soup.select('div.meta-value.genre')
    gen = str(gres).replace(' ','').replace('<divclass="meta-valuegenre"data-qa="movie-info-item-value">','').replace('</div>','').replace('\n','')
    des=soup.select('p.mop-ratings-wrap__text.mop-ratings-wrap__text--concensus')
    des=str(des[0]).replace('<p class="mop-ratings-wrap__text mop-ratings-wrap__text--concensus">','').replace('<em>','').replace('</em>','').replace('</p>','')
    genres.append(gen)    
    desc.append(des)
i=0
k=1
g=0
while(i<len(rt)):
    main.append(["rotten tomatoes",k,rt[i+2],str(rt[i+1]),desc[g],genres[g]])
    i+=4
    k+=1
    g+=1
print("Done")  
```

```python
#getting genre dictionary and top 50 review words sorted according to frequency for imdb website 
import string
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
imdb=[]
c=""
for i in range(0,50):
    data={
        main[i][2]:main[i][5],
    }
    imdb.append(data)
    c=c+" "+main[i][4]
    
c=c.translate(str.maketrans('', '', string.punctuation))
c = [word for word in c.split() if word not in (stopwords.words('english'))]
d = {}
for word in c: 
    d[word] = d.get(word, 0) + 1
word_freq_imdb = []
for key, value in d.items():
    word_freq_imdb.append((value, key))
word_freq_imdb.sort(reverse=True) 
word_freq_imdb=word_freq_imdb[0:50]
print("TOP 50 USED WORDS")
print(word_freq_imdb)
```

```python
#getting genre dictionary and top 50 review words sorted according to frequency for Rotten tomatoes website 
rt=[]
c=""
for i in range(100,150):
    data={
        main[i][2]:main[i][5],
    }
    rt.append(data)
    c=c+" "+main[i][4]
    
c=c.translate(str.maketrans('', '', string.punctuation)).lower()
c = [word for word in c.split() if word not in (stopwords.words('english'))]
d = {}
for word in c: 
    d[word] = d.get(word, 0) + 1
word_freq_rt = []
for key, value in d.items():
    word_freq_rt.append((value, key))
word_freq_rt.sort(reverse=True) 
word_freq_rt=word_freq_rt[0:50]
print("TOP 50 USED WORDS")
print(word_freq_rt)
```
